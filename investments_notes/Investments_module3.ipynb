# Module 3: Capital Asset Pricing Model (CAPM)

## CAPM 

* CAPM Equation: $E(r_{i}) = r_{f} + \beta_{i}[E(r_{m}) - r_{f}]$

    * CAPM equation shows the relation between expected return and risk

    * CAPM says for any firm _i_ that is exposed to a certain level of systematic risk $\beta_{i}$, then a fair return is given by $E(r_{i})$

* CAPM is a set of predictions concerning equilibrium expected returns on risky assets

* Assumptions of CAPM:
    - Individual Behavior
        - Investors are rational, mean-variance optimizers
        - Their common planning horizon is a single period
        - Investors all use identical input lists, an assumption often termed homogenous expectations; all relevant information is publically available

    - Market Structure
        - All assets are publicaly held and trade on public exchanges
        - Investors can borrow or lend at a common risk-free rate, and they can take short positions on traded securities
        - No Taxes, nor transaction costs


## The Security Market Line

* Security Market Line Equation: $E(r_{M}) = r_{f} + \beta_{M}[E(r_{M}) - r_{f}]$
* By definition, the systematic risk of the market to the market is 1, there is a 1-to-1 relation b/w the two; $\beta_{M} = 1$
* SML helps us identify securities that are mispriced; theoretically all traded securities should be somewhere on the SML based on their $\beta$, their risk relative to the market

##### Positive-Alpha

* If a security is higher than the SML; it's return relative to it's $\beta$ is higher than the SML, we can say the security is undervalued
* The y distance between the the security and the SML is $\alpha$





## Valuation

* Book Value based on historical costs
* Market values measure the current values of assets and liabilities and will probably not match Book Value

#### Dividend Discount Model

* Calculate the value of a firm that pays dividends by discounting future dividends into perpetuity

    * $V_{0} = \frac{D_{1}}{1+k} + \frac{D_{2}}{1+k} + ... + \frac{D_{n}}{1+k}$

* For dividends that grow constantly:
    * $V_{0} = \frac{D_{0}(1+g)}{k-g} = \frac{D_{1}}{k-g}$

* Example; Constang Growth DDM:

    - A stock just paid an annual dividend of $3/share
    - Dividend is expected to grow at 8% indefinitely
    - Market capitalization rate is 14%

    - $V_{0} = \frac{D_{1}}{k-g} = \frac{3 * 1.08}{.14-.08} = \$54$

    - If dividends do not grow, then set g to 0 to recalculate

        - $V_{0} = \frac{D_{0}}{k} = \frac{3}{.14}$


* Implications of Dividend Discount Model:
    - The stock's value will be greater the larger it's expected dividend per share
    - The stock's value will be greater the lower the market capitalization rate, _k_
    - The stock's value will be greater the higher the expected growth rate of dividends

    - The stock price is expected to grow the same rate as the dividend growth rate

##### Estimating Dividend Growth Rates

* Dividend Growth Rate: $g = ROE * b$
    - $g$ = growth rate of dividends
    - $ROE$ = Return on Equity
    - $b$ = plowback or retention rate (b = 1-dividend payout ratio)
    - This works best when a firm achieves a steady ROE, if ROE is unsustainably high, the formula doesn't work

#### Present Value of Growth Opportunities
* The value of the firm equals:
    - The value of the assets already in place +
    - The NPV of its future investments or the PV of growth opportunities
    - Price/Share = No-growth value/share + PV Growth Opportunity
    - $P_{0} = \frac{E_{1}}{k} + PVGO$    

* Example: Poor Reinvestment
    - Earnings of \$5/share
    - Firm reinvests 60% of its earnings
    - Expected year-end dividend is \$2/share
    - Project generates ROE of 10%
    - Capitalization rate is 15% 
        - Project ROE is less than the required return on capital
        - Capitalization rate is used as the discount rate
        - Suggests that the company's internal opportunities for return are not very good compared to the rest of the market
    - $g = ROE * b = 10\% * .6 = 6%$
    - $P_{0} = \frac{\$2}{15\% - 6\%} = \$22.22$

    - PVGO = Price / share - no-growth value/share
        - $PVGO = \$22.22 - \frac{5}{.15} = -\$11.11$
        - PVGO is negative because the firm insists on retaining all earnings $5 and investing in projects with lower return than the cost of capital; it should just pay out all earnings $5 as dividends to raise the stock valuation

* Two-Stage Dividend Discount Model
    - Stage 1 involves discounting individual dividend payments to present value
    - Stage 2 involves discounting perpetual future dividends to present value
    - $V_{0}$ = current value
    - g = terminal (long-term) dividend growth rate
    - k = required rate of return
        - k can be estimated as the CAPM expected return
    
    - $V_{0} = \frac{D_{1}}{(1+k)^{1}} + \frac{D_{2}}{(1+k)^{2}} + \frac{D_{3}}{(1+k)^{3}} + \frac{D_{4}}{(k-g)(1+k)^{3}}$

        - Stage 2 part of the formula ($\frac{D_{4}}{(k-g)(1+k)^{3}}$) is a pertuity formula
        - the discounting of $D_{4}$ by $k-g$ gets you the discounted values of perpetual dividends at time = 3
        - the second part $(1+k)^{3}$ discounts the perpetuity back to time = 0







#### Comparables (Ratio) Valuation

* Ratios can be used in valuation
* Many valuation ratios **DO NOT** represent an independent valutation of a firm
    - Ratios such as "price-to-earnings" ratios incorporate the market stock price; the ratio includes a market valuation and cannot be used for independent valuation
    - Independent valuation is attempting to find the intrinsic price that is most likely not the same as the current market price
* Example: Valuation Using Forward P/E Ratio
    - Forecasted earnings per share (EPS) = \$3.45
    - Median Industry P/E Ratio = 12
    - Then, $P_{0} = EPS * 12 = \$41.40$
    - Forward P/E ratio: using one-year-ahead forecast of earnings
    


##### Cyclically Adjusted P/E Ratio (CAPE)

* A Cyclically Adjusted P/E ratio to avoid the problems associated with using P/E ratios over different phases of the business cycle
* Divide Share Price by an estimate of sustainable long-term earnings rather than current earnings
    - such estimate could be an average inflation-adjusted earnings over an extended period of time (i.e. 10 years)

##### Free Cash Flows to the Firm (FCFF)

* Free Cash Flows to the Firm (FCFF) is the after-tax cash flow generated by the firm's operations, net of investments in fixed assets and changes in working capital
* FCFF represents the cash flow available to all stakeholders
* Value the firm by discounting FCFF using the _weighted-average cost of capital (WACC)_

###### Weighted Average Cost of Capital (WACC)

* $WACC = \frac{MV_{E}}{MV_{E} + MV_{D}}k_{E} + \frac{MV_{E}}{MV_{E} + MV_{D}}k_{D}(1-t)$

###### Enterprise Value

* Enterprise Value = $\sum^{T}_{t=1} \frac{FCFF_{t}}{(1+WACC)^{t}} + \frac{V_{T}}{(1+WACC)^{T}}$
    - Where $V_{T} = \frac{FCFF_{T+1}}{WACC - g}$

###### Free Cash Flows to the Firm Calculation

* FCFF = After tax EBIT + depreciation - CapEX - Change in net working capital

* $FCFF = EBIT * (1-t) + Depreciation- CapEX - \Delta NWC$

##### Free Cash Flow to Equity (FCFE)

* Free cash flow to equity, FCFE, is obtained by removing from FCFF all cash flows associated with debtholders
* Resulting measure is cash flows associated with equity holders
* $FCFE = FCFF - Interest * (1-t) + \Delta Debt$
* Intrinsic Value of Equity = $\sum^{T}_{t=1} \frac{FCFE_{t}}{(1+k_{E})^{t}} + \frac{E_{T}}{((1+k_{g})^{T})}$

    - Where $E_{T} = \frac{FCFE_{T+1}}{k_{E}-g}$
    - and $k_{E}$ may be obtained using CAPM
