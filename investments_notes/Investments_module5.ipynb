## Efficient Market Hypothesis

* The theory that stocks reflect all available information
* Reflevent information and news is considered random and unpredictable, and as stock prices reflect available information, stock price changes follow a random walk; stock prices go up and down regardless of past performance

### Competition as the Source of Efficiency

* Information is the most precisous financial commodity
* Analysts have an incentive to discover information and use it to make money
* Strong competition assures prices reflect all information
* Once prices reflect information, it becomes more difficult to make money

### Efficiency Differs Across Markets

* Small stocks receive less analyst coverage than loarge stocks in the US
* Emerging markets are less intensely analyzed than US markets and may have less rigourous disclosure requirements

### Inefficiency's Negative Impacts

* Inefficient markets imply there is systematic resource misallocation
    - Firms with overvalued securities can raise capital too cheaply
    - Firms with undervalued securities may pass up profitable opportunities because their cost of capital is unfairly high

### Event Studies

* Financial research can assess the impact of an event on a firm's stock price
* The abnormal return due to the event is the difference between the stock's actual return and a proxy for the stock's return in the absence of the event

#### Test Structure

* Returns are examined to determine if they are abnormal compared to what they would have been without the event
    - Proxy for normal returns in the absence of the even can be the market model (CAPM)
    - $r_{t} = \alpha + \beta r_{Mt} + e_{t}$
        - $\alpha + \beta r_{Mt}$ represents the predicted returns according to the market model without the event (CAPM)
        - $e_{t}$ is a residual figure and represents the effect of the event; we can rearrange the market model to solve for $e_{t}$
            - $e_{t} = r_{t} - (\alpha + \beta r_{Mt})$

### Types of Stock Analysis

* Technical Analysis: search for patterns in stock prices and trading volumes. The patterns are used to predict price movement
    - Successful technical analysis depends on a sluggish response of stock prices to supply-demand factors
* Fundamental Analysis: Uses earnings and dividend prospects of the firm to determine proper (or intrinsic) stock prices
    - Analysis seeks firms that are mispriced

### Versions of Efficient Market Hypothesis

* Three levels of efficiency within the Efficient Market Hypothesis to reflect differing definitions of _all available information_; all versions assert that prices reflect __all available information__
    - Weak
    - Semi-strong
    - Strong

#### EMH Weak Form

* Weak-form assets that stock prices already reflect all information that can be derived by examining market trading data such as historical pricing, trading volumes, or short interest

* As all relevent information used for technical analysis is already priced in, weak-form efficiency implies that __technical analysis is useless__

#### EMH Semi-strong form

* Semi-strong states that all publicaly available information regarding the prospects of a firm are already reflected in stock prices

* Information include past prices, fundemental data on firm's product line, quality of management, balance sheet composition, patents held, earnings forecasts, dividend prospects, and accounting practices

* Implies that fundemental and technical analysis is useless

#### EMH Strong Form

* Strong-form states that all relevant information to the firm is already reflected in the stock price; even insider information

* Implies that insider information, technical and fundemental analysis is useless
    - This is hard to believe

### Implications of EMH on Management Strategies

* Efficient Market Hypothesis implies that the costs of research are too onerous for small players, only large portfolios can afford it
* Active Portfolio Management is an expensive strategy that takes significant resources to be successful
    - Clients pay high fees to cover costs of resources and trading (analyst time, data, trading comissions)
    - Must beat markets consistently to be worthwhile
    - Suitable for very large portfolios where analysts have a comparative advantage
    - Frequent selling may triger greater tax liabilities 

* Passive Portfolio Management does not attempt to beat the market; just accepts market returns
    - Very low costs (low fees for investors) as few resources needed to manage passive portfolios

* EMH implies that passive investment strategy is the most efficient
    - There is a little point in paying high fees for analysis that is unlikely to lead to superior performance


## Behavioral Finance and Technical Analysis

### Conventional vs Behavioral Finance

* Conventional Finance:
    - Prices are correct and qual to instrinsic value
    - Resources are allocated efficiently
    - Consistent with EMH

* Behavioral Finance: 
    - Assumes investors do not behave rationally
    - Arbitrageurs are limited and therefore insufficient to force prices to match intrinsic value

#### Two Categories of Irrationalities

* Information Processing Errors: Investors do not always process information correctly; as such investors form incorrect opinions of probability distributions for future returns
    - Limited attention span, underreaction, and overreaction; reliance on heuristics due to limited time/attention
    - Overconfidence; people tend to overestimate the precision of their beliefs or forecasts, and they tend to overestimate their abilities
    - Extrapolation, pattern recognition, representative bias; Individuals are adept at discerning patterns, even percieving patterns that don't actually exist

* Behavioral Biases: Even when given a correct probability distribution of returns, investors may make inconsistent or suboptimal decisions
    - Framing: Decisions affected by how choices are described, such as whether uncertainty is posed as potential gains from a low baseline livel, or as losses from a higher baseline level
    - Mental accounting: Specific form of framing in which people segregate certain decisions
    - Regret avoidance: Individuals who make decisions that turn out badly have more regret when that decision was more unconventional
    - Affect and feelings: Investors tend to choose stocks with higher affect (i.e. stocks they admire) driving prices up and returns down

#### Limits to Arbitrage

* Behavioral biases would not matter if rational arbitrageurs could fully exploit the mistakes of behavioral investors; bringing the price back to rational levels

    - The problem with this statement is that markets can remain irrational longer than the arbitrageurs can remain solvent

* Behavioral Finance often explains bubbles as a result of overconfidence and representativeness biases
